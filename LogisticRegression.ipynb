```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python
# 讀取資料
url="https://vincentarelbundock.github.io/Rdatasets/csv/carData/TitanicSurvival.csv"
titanic =pd.read_csv(url, index_col=0)
```

```python
# 將 Age 遺漏值以 median中數 填補
age_median=np.nanmedian(titanic["age"])
new_age=np.where(titanic["age"].isnull(), age_median, titanic["age"])
titanic["age"] =new_age
titanic
```

```python
# 創造dummy variables （補值）
titanic[['passengerClass', 'sex', 'survived']] =titanic[['passengerClass', 'sex', 'survived']].apply(LabelEncoder().fit_transform)
```

```python
features =titanic[['passengerClass', 'sex', 'age']]
target =titanic['survived']
```

```python
# 建立模型
logistic = LogisticRegression()
logistic.fit(features, target)

# 印出係數
print(logistic.coef_)

# 印出截距
print(logistic.intercept_)
```
